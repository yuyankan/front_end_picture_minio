```python
import myquery_db as query_db
import pandas as pd
import datetime
```

```python
table =  {'raw_data':'ks_project_yyk.EHS.ods_raw_data' }
```

```python
def get_sqldata(production_line= 'C11',cameraid=4):
    utcnow = datetime.datetime.utcnow()-datetime.timedelta(days=1)
    utcday = utcnow.strftime('%Y-%m-%d')
    year = utcnow.year
    quarter = (utcnow.month - 1) // 3 + 1   # Q1~Q4

    partition_key = f"{production_line}_{cameraid}_{year}{quarter}"

    query = f'''
        select *
        from {table['raw_data']}
        where 1=1
        and partition_key = '{partition_key}'
        and createtime_utc >= '{utcday}'
        '''
    print(query)
    df_raw = query_db.query_ksdata(query)
    latest_image_url = None
    cur_time = None

    if not df_raw.empty:
        print(f"No data found for partition_key: {partition_key} on {utcday}")
        df_raw['createtime_cn'] = df_raw['createtime_utc'].dt.tz_localize(None)+ pd.Timedelta(hours=8)
        df_raw['createtime_cn'] = df_raw['createtime_cn'].dt.strftime('%Y-%m-%d %H:%M:%S')
        latest_image_url, cur_time = df_raw.sort_values(by='createtime_cn', ascending=False).iloc[0]['object_name']
    return df_raw, latest_image_url,cur_time
```

```python
df_raw, latest_image_url= get_sqldata()
```
